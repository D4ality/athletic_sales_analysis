```python
# Import Libraries and Dependencies
import pandas as pd
import numpy as np
```

### 1. Combine and Clean the Data
#### Import CSVs

```python
# Read the CSV files into DataFrames.
athletic_sales_2020_df = pd.read_csv('athletic_sales_2020.csv')
athletic_sales_2021_df = pd.read_csv('athletic_sales_2021.csv')
```

```python
# Display the 2020 sales DataFrame
athletic_sales_2020_df.head(15)
```

```python
# Display the 2021 sales DataFrame
athletic_sales_2021_df.head(15)
```

#### Check the data types of each DataFrame

```python
# Check the 2020 sales data types.
print(athletic_sales_2020_df.dtypes)
```

```python
# Check the 2021 sales data types.
print(athletic_sales_2021_df.dtypes)
```

#### Combine the sales data by rows.

```python
# Combine the 2020 and 2021 sales DataFrames on the rows and reset the index.
combined_data_rows_df = pd.concat([athletic_sales_2020_df, athletic_sales_2021_df], axis="rows", ignore_index=True)
                            


combined_data_rows_df.head(15)
```

```python
# Check if any values are null.
print(combined_data_rows_df.isnull().sum())
```

```python
# Check the data type of each column
print(combined_data_rows_df.columns)
```

```python
# Convert the "invoice_date" to a datetime datatype
combined_data_rows_df['invoice_date'] = pd.to_datetime(combined_data_rows_df['invoice_date'])
```

```python
# Confirm that the "invoice_date" data type has been changed.
print(combined_data_rows_df.dtypes)
```

### 2. Determine which Region Sold the Most Products

#### Using `groupby`

```python
# Show the number products sold for region, state, and city.
# Rename the sum to "Total_Products_Sold".

df = combined_data_rows_df.groupby(['region', 'state', 'city'])['units_sold'].agg(Total_Products_Sold=("sum"))


# Show the top 5 results.
df.sort_values(by=(['Total_Products_Sold']), ascending=False).head()                       
```

#### Using `pivot_table`

```python
# Show the number products sold for region, state, and city.
total_products_sold = pd.pivot_table(combined_data_rows_df, 
                                     index= ['region', 'state', 'city'],
                                     values='units_sold',
                                     aggfunc='sum')

# Rename the "units_sold" column to "Total_Products_Sold"
total_prdcts = total_products_sold.rename(columns={"units_sold": "Total_Products_Sold"})

# Show the top 5 results.
total_prdcts.sort_values(by=(['Total_Products_Sold']), ascending=False).head()  
```
